Applying decision tree classification and finding confusion matrix as well as random forest classifier and finding accuracy on horse.csv dataset

```python
import pandas as pd
```

```python
df=pd.read_csv("horse.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df
```

```python
df.columns
```

```python
# categorical="surgery",age,temp_of_extremities,periheralpulse,mucous_membrane,capillar,abdomo-appeara,outcome,surgicallession,cpdata
df_categorical=df[['surgery', 'age','temp_of_extremities', 'peripheral_pulse','pain', 'peristalsis','abdominal_distention','nasogastric_tube','rectal_exam_feces', 'abdomen', 'nasogastric_reflux','mucous_membrane', 'capillary_refill_time',
                   'abdomo_appearance','outcome', 'surgical_lesion','cp_data']]
df_numerical=df[['hospital_number', 'rectal_temp', 'pulse','respiratory_rate',
                 'nasogastric_reflux_ph','packed_cell_volume', 
                 'total_protein','abdomo_protein','lesion_1', 'lesion_2','lesion_3']]
```

```python
df_categorical_encoded = pd.get_dummies(df_categorical, drop_first=True)
df_categorical_encoded.head()
```

```python
df_new = pd.concat([df_numerical, df_categorical_encoded], axis=1)
df_new.head()
```

```python
df_new.columns
```

```python
df_new.shape
```

```python
x=df_new.iloc[:,1:64].values
y=df_new["cp_data_yes"].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion="gini")
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(x_train, y_train)

```

```python
y_pred = clf.predict(x_test)
print(y_pred)
```

```python

from sklearn import metrics
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```
